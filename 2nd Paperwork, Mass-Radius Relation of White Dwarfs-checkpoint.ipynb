```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

mu = eval(input("Average nucleus of free electron:"))
r_array = []
m_array = []

def derivatives(f, t, epsilon=1e-8):
    y, z = f
    if z < 0:
        dydt = 0
        dzdt = 6 * np.pi**2 * y / t**2
        return dydt, dzdt

    dydt = -t**2 * z**1.5
    if t < epsilon:
        dzdt = -2 * np.pi**2 * np.sqrt(1 + z) * z**1.5 * t
    else:
        dzdt = 6 * np.pi**2 * y * np.sqrt(1 + z) / t**2
    return dydt, dzdt

r_0 = 2.11 * 1e9 * (mu / 2) ** (-1)
rho_c_array = 10 ** np.linspace(5, 11, 500)
rho_0 = 1.9479 * 1e6 * (mu / 2)
xc_array = (rho_c_array / rho_0) ** (2 / 3)

for xc in xc_array:
    f0 = [0, xc ** 2]
    t_array = np.linspace(0, 1, 500)
    f_array = odeint(derivatives, f0, t_array)

    y_array = f_array[:, 0]
    z_array = f_array[:, 1]

    zero_index = np.where(z_array <= 0)[0]
    if zero_index.size > 0:
        y_at_zero_z = np.abs(y_array[zero_index[0]])
        t_at_zero_z = t_array[zero_index[0]]
        r_at_zero_z = r_0 * t_at_zero_z
        r_array.append(r_at_zero_z)
        m_i = 4 * np.pi * rho_0 * (r_at_zero_z / t_at_zero_z) ** 3 * y_at_zero_z * 1e-33
        m_array.append(m_i)

plt.semilogx(r_array, m_array, label='Numerical result')

r_theory = np.linspace(min(r_array), max(r_array), 500)
m_non_relativistic = 1 / r_theory**3
plt.semilogx(r_theory, m_non_relativistic, label='Non-relativistic limit', linestyle='--')

chandrasekhar_mass = 1.44 * 1.989e30 / 1e30
m_relativistic = [chandrasekhar_mass] * len(r_theory)
plt.semilogx(r_theory, m_relativistic, label='Relativistic limit (Chandrasekhar)', linestyle='--')

measured_masses = [0.602, 1.034, 0.43, 0.48, 0.75, 0.59, 0.70, 0.53, 0.46, 0.50, 0.59]
measured_radii = [0.01234, 0.0074, 0.0124, 0.0111, 0.0084, 0.0150, 0.0149, 0.0120, 0.013, 0.011, 0.011]

M_sun = 1.989e30
R_sun = 6.955e10

measured_masses = np.array(measured_masses) * M_sun / 1e30
measured_radii = np.array(measured_radii) * R_sun

plt.scatter(measured_radii, measured_masses, color='red', label='Measured data')

plt.xlabel('Radius (log scale)')
plt.ylabel('Mass / 10^30kg')
plt.legend()
plt.show()
```
